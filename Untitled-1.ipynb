采样频率 $ F_{samlping} = $ `44100`

***

如若需要合成的声音,

频率为 $ f $ `(Hz)`,

则其所对应的

周期 $ T = \frac{1}f $ `(s)`,

因此一块噪声的长度 $ p = F_{sampling} \cdot T = \frac{F_{sampling}}f $,

***

注:
* *`Karplus-Strong` 算法在 `[300, 1000] (Hz)` 范围内表现较好。*

```python
import numpy as np
import scipy.io as scio
from scipy.io import wavfile
import scipy.signal as scisig
from matplotlib import pyplot as plt
plt.style.use("solarized-light")

class Constants(object):
    FLAG_MAIN = True # 运行程序的开关
    FREQ_MIDDLE_C = 261.6255653 # 中央C的频率
    FREQ_SAMPLING_RATE = 44100 # 采样频率
    NUM_CHANNELS = 1 # 设为单声道
    COEF_DELAY = 1.28 # 延音系数
    COEF_SPEED = 192 / 120 # 速度系数
    PATH_MATMIDI = "./STAGE_OF_SEKAI.mat" # 存储 midi 的`.mat`张量的索引 { 0: 频率, 1: 起始时刻, 2: 结束时刻 }
    LEN_NOISECHUNK = 3939
    NOISECHUNK = np.random.random((LEN_NOISECHUNK, NUM_CHANNELS)) # 一块基础噪声, 其每个第一个周期的内容决定着音色, 其长度足够即可, 保证其长度 > (采样率 / 频率) + 1
Constants.NOISECHUNK -= Constants.NOISECHUNK.mean()

midi2freq = lambda x : 55 * (2 ** (x / 12 - 2.75)) # (440 / 32) * (2 ** (x - 9) / 12)
```

```python
def karplus_strong(freq = Constants.FREQ_MIDDLE_C, ti = 3.9):

    len_period = np.ceil(Constants.FREQ_SAMPLING_RATE / freq).astype(int) # 80

    #noisechunk = Constants.NOISECHUNK[ : len_period , : ].copy()
    noisechunk = Constants.NOISECHUNK[ : : Constants.LEN_NOISECHUNK // len_period , : ].copy()[ : len_period , : ]

    len_output = np.round(Constants.FREQ_SAMPLING_RATE * ti).astype(int)
    num_periods = int(len_output / len_period) - 1

    new_indices = [noisechunk.shape[0] - 1] + list(range(noisechunk.shape[0] - 1))

    res = np.zeros((len_output, Constants.NUM_CHANNELS))
    for k in range(num_periods):
        noisechunk = (noisechunk + noisechunk[new_indices, : ]) / 2
        res[ len_period * k : len_period * (k + 1) , : ] += noisechunk

    return res
```

```python
def parsemidi(filepath = Constants.PATH_MATMIDI, keyname = "arr"):
    mat = scio.loadmat(filepath)[keyname]

    len_output = np.ceil((mat[ -1, 2 ] + 3.9) * Constants.FREQ_SAMPLING_RATE / Constants.COEF_SPEED).astype(int) # +16.4 是为了防止音符渲染出的音频溢出总音频长度
    res = np.zeros((len_output, Constants.NUM_CHANNELS))

    for eachnote in list(mat):
        freq_midi    = midi2freq(eachnote[0])
        time_starter = eachnote[1]
        time_ender   = eachnote[2]
        index_starter = int(round(time_starter * Constants.FREQ_SAMPLING_RATE / Constants.COEF_SPEED)) # 加入速度系数
        newnote = karplus_strong(freq_midi, (time_ender - time_starter) * Constants.COEF_DELAY / Constants.COEF_SPEED)
        res[ index_starter : index_starter + newnote.shape[0] , : ] += newnote

    res /= np.abs(res).max()

    return res
```

```python
if Constants.FLAG_MAIN:
    scio.wavfile.write(
        "./STAGE_OF_SEKAI.wav",
        Constants.FREQ_SAMPLING_RATE,
        parsemidi()
        )
```

```python
arr2plot = karplus_strong()
idx = list(range(arr2plot.shape[0]))
plt.rcParams["figure.figsize"] = (16, 9)
plt.stem(idx[ : : 4410 ], arr2plot[ : : 4410 , : ])
plt.plot(idx[ : : 4410 ], arr2plot[ : : 4410 , : ])
plt.stem(idx[ : : 3900 ], arr2plot[ : : 3900 , : ])
plt.plot(idx[ : : 3900 ], arr2plot[ : : 3900 , : ])
plt.title("Karplus-Strong")
plt.savefig("./cover.jpeg")
plt.show()
```
