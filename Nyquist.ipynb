```python
import numpy as np
import tqdm

from matplotlib import pyplot as plt
plt.style.use("solarized-light")
```

```python
xd = np.concatenate([ np.arange(0, 2 * np.pi, np.pi / 2), [2 * np.pi] ])
xs = np.concatenate([ np.arange(0, 2 * np.pi, np.pi / 128), [2 * np.pi] ])
freqx2pi = np.arange(1, 15, 4)
```

```python
with plt.rc_context({"figure.figsize": [32, 5.6]}):
    plt.stem(xd, np.sin(xd))
    for k in tqdm.tqdm(freqx2pi):
        plt.plot(xs, np.sin(k * xs), "--" if k >= 8 else "-", label="freq= %d/2pi Hz" % k)
    plt.plot([ np.pi * (1 - 1/k) / 2 for k in tqdm.tqdm(freqx2pi) ], [0] * len(freqx2pi), "o", label="1/8 Term before the 2nd Sample Point")
    plt.title("Signals of Different Frequencies at the same Sample Rate")
    plt.xlabel("Time [s]")
    plt.ylabel("Amplitude")
    plt.legend()
    if True:
        plt.savefig("./Nyquist.jpeg")
    plt.show()
```
